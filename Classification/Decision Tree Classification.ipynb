```python
x =  churn.drop(['Churn']  , axis  =  1)
y  = churn['Churn']
```

```python
### 
```

```python
y.value_counts()
```

```python
# create pipeline
```

```python
cat  =  list(x.columns[x.dtypes =='object'])
num  =  list(x.columns[x.dtypes != 'object'])
```

```python
cat
```

```python
x[cat]
```

```python
x['TotalCharges'].dtype
```

```python
x['TotalCharges']  = pd.to_numeric(x['TotalCharges'] ,errors='coerce')
```

```python
x[num].head(2)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OrdinalEncoder , StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline  =  Pipeline(steps  = [('scaler' , StandardScaler())])
cat_pipeline  = Pipeline(steps = [('encoder' , OrdinalEncoder( ))])
```

```python
pre =ColumnTransformer([('num_pipeline' , num_pipeline , num) , ('cat_pipeline' , cat_pipeline  , cat)])
```

```python
pre
```

```python
x1 = pre.fit_transform(x)
```

```python
x1  =  pd.DataFrame(x1 , columns  =  pre.get_feature_names_out())
```

```python
x1.head()
```

```python
y
```

```python
y.replace({'Yes':1 , "No":0} , inplace  =  True)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
# train test split 
x_train , x_test , y_train , y_test  =  train_test_split(x1, y , test_size  = 0.2 , random_state  =  34)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# model fitting 

from sklearn.tree import DecisionTreeClassifier
model  =   DecisionTreeClassifier(criterion  =  'gini' , max_depth  =  6)
```

```python
model.fit(x_train , y_train)
```

```python
model.score(x_train , y_train )
```

```python
# training Evaluation
y_pred  =  model.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score , precision_score ,\
recall_score, \
f1_score , classification_report , confusion_matrix
```

```python
confusion_matrix(y_train , y_pred)
```

```python
acc  =  accuracy_score(y_train, y_pred)
precision = precision_score(y_train , y_pred )
recall  = recall_score(y_train , y_pred )
f1 =  f1_score(y_train , y_pred )  

print(f"acc --> {acc}")
print(f"precision --> {precision}")
print(f"recall --> {recall}")
print(f"f1 --> {f1}")
```

# testing Evaluation

```python
y_test_pred  =  model.predict(x_test)
```

```python
acc  =  accuracy_score(y_test, y_test_pred)
precision = precision_score(y_test, y_test_pred)
recall  = recall_score(y_test, y_test_pred)
f1 =  f1_score(y_test, y_test_pred) 

print(f"acc --> {acc}")
print(f"precision --> {precision}")
print(f"recall --> {recall}")
print(f"f1 --> {f1}")
```

```python
print(classification_report(y_test , y_test_pred))
```
