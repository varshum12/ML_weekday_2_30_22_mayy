## Seperate x and y variable

```python
df.columns
```

```python
df.drop(['id']  , axis  = 1, inplace = True)
```

```python
x  =  df.drop(['Price'] ,axis = 1)
y = df['Price']
```

##  Seperate numeric and categorical data

```python
cat = []
num = []
for i in x.columns:
    if x[i].dtype == 'object':
        cat.append(i)
    else:
        num.append(i)
```

```python
cat
```

```python
num
```

# For Creation of pipeline

```python
from sklearn.preprocessing import OneHotEncoder , StandardScaler
from sklearn.pipeline import  Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline  =  Pipeline(steps= [('Scaler'  , StandardScaler()) ,
                                   ('imputer' , SimpleImputer(strategy= 'mean'))])
cat_pipeline  =  Pipeline(steps  = [('encoder' ,  
                                     OneHotEncoder(handle_unknown= 'error'))   , ('imputer' , SimpleImputer(strategy= 'most_frequent'))])
```

```python
pre =  ColumnTransformer([('num_pipeline'  , num_pipeline  , num)  , 
                          ('cat_pipeline' , cat_pipeline , cat)])
```

```python
pre
```

```python
cols  =  x.columns
cols
```

```python
x1 = pre.fit_transform(x)
```

```python
x1
```

```python
x1  =  pd.DataFrame(x1.toarray() , columns= pre.get_feature_names_out())
```

```python
x1
```

```python
N= x1.isnull().sum()
```

```python
N[N>0]
```

# Split train - test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train  , x_test , y_train , y_test =  train_test_split(x1 , y , test_size= 0.15 , random_state= 7)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_test.shape)
print(y_train.shape)

```

# Model Fitting

```python
from sklearn.linear_model import LinearRegression
LR  = LinearRegression()
```

```python
LR.fit(x_train , y_train )
```

```python
LR.intercept_
```

```python
LR.coef_
```

# EVALUATE TRAINING MODEL

```python
y_train_pred =  LR.predict(x_train)
```

```python
from sklearn.metrics import mean_squared_error , root_mean_squared_error , mean_absolute_error , mean_absolute_percentage_error , r2_score
```

```python
MSE  =  mean_squared_error(y_train  , y_train_pred)
RMSE  =root_mean_squared_error(y_train  , y_train_pred)
MAE = mean_absolute_error(y_train  , y_train_pred)
MAPE  =  mean_absolute_percentage_error(y_train  , y_train_pred)
r2 = r2_score(y_train  , y_train_pred)

print(f'MSE --> {MSE} ')
print(f'RMSE --> {RMSE} ')
print(f'MAE --> {MAE} ')
print(f'MAPE --> {MAPE} ')
print(f'R2 --> {r2} ')
```
