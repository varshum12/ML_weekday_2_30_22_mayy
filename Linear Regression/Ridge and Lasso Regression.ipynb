```python
# seperate x and y 
x  =  df.drop(['Price'] , axis =  1)
y =  df['Price']
```

```python
# Use pipeline for preprocessing
from sklearn.preprocessing import OrdinalEncoder  , StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OrdinalEncoder , StandardScaler
from sklearn.impute import SimpleImputer
```

```python
cat  =  []
num  = []
for i in x.columns:
    if x[i].dtype == 'object':
        cat.append(i)
    else:
        num.append(i)
```

```python
# Seperate numerical  data  
num  =[]
cat  = []
for i in x.columns:
    if x[i].dtype!='object':
        num.append(i)
    else:
        cat.append(i)
```

```python
print(cat)

print()

print(num)
```

```python

# create num and cat pipeline
num_pipeline  =  Pipeline(steps =  [('imputer'  , SimpleImputer(strategy= 'mean' )),
                                          ('scaler' ,StandardScaler()) ] )

cat_pipeline  =  Pipeline(steps  =  [('encoder' , OrdinalEncoder()) ,
                                    ('imputer'  , SimpleImputer(strategy= 'most_frequent'))])                                          
```

```python
# connect pipeline
pre  = ColumnTransformer([('num_pipeline' , num_pipeline , num)  , 
                         ('cat_pipeline' , cat_pipeline , cat )])
```

```python
pre
```

```python
# Apply to x variables
```

```python
x
```

```python
x1 = pre.fit_transform(x)
```

```python
x1
```

```python
# convert in dataframe

x1  =  pd.DataFrame(x1 , columns = pre.get_feature_names_out())
```

```python
x1
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test  , y_train , y_test = train_test_split(x1  , y , test_size  = 0.15 , random_state = 12)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
R   = Ridge(alpha=75.0 , max_iter= 5)
```

```python
# Model fitting
```

```python
R.fit(x_train  , y_train )
```

### Training Evaluation

```python
y_train_pred = R.predict(x_train)
```

```python
from sklearn.metrics import mean_absolute_error ,mean_squared_error,\
mean_absolute_percentage_error ,r2_score  , root_mean_squared_error
```

```python
MSE  =  mean_squared_error(y_train  , y_train_pred)
RMSE  =root_mean_squared_error(y_train  , y_train_pred)
MAE = mean_absolute_error(y_train  , y_train_pred)
MAPE  =  mean_absolute_percentage_error(y_train  , y_train_pred)
r2 = r2_score(y_train  , y_train_pred)

print(f'MSE --> {MSE} ')
print(f'RMSE --> {RMSE} ')
print(f'MAE --> {MAE} ')
print(f'MAPE --> {MAPE} ')
print(f'R2 --> {r2} ')
```

```python
# Test Evaluation
y_test_pred = R.predict(x_test)
MSE  =  mean_squared_error(y_test  , y_test_pred)
RMSE  =root_mean_squared_error(y_test  , y_test_pred)
MAE = mean_absolute_error(y_test  , y_test_pred)
MAPE  =  mean_absolute_percentage_error(y_test  , y_test_pred)
r2 = r2_score(y_test  , y_test_pred)

print(f'MSE --> {MSE} ')
print(f'RMSE --> {RMSE} ')
print(f'MAE --> {MAE} ')
print(f'MAPE --> {MAPE} ')
print(f'R2 --> {r2} ')
```
